```python
import codecs, base64, binascii

data = "fns_pgs{83a145c0-8625-408d-986c-938cf95b7d9a}"

# ROT13
print("ROT13:", codecs.encode(data, "rot_13"))

# Base64 (try decoding, ignoring errors)
try:
    print("Base64:", base64.b64decode(data).decode())
except Exception:
    print("Base64: not valid")

# Hex
try:
    print("Hex:", binascii.unhexlify(data).decode())
except Exception:
    print("Hex: not valid")
```
